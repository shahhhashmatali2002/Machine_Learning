```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
df_n = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
df_n
```

```python
df_n.isnull().sum()
```

```python
import math
median_Age = math.floor(df_n.Age.median())
```

```python
Age_n = df['Age'].fillna(median_Age)
new_df = df_n.drop(['Age'],axis='columns')
new_df 
```

```python
dummies = pd.get_dummies(new_df.Sex)
dummies.head()
```

```python
sex_n = []
for i in new_df.Sex:
    if i == 'male':
        sex_n.append(0)
    elif i == 'female':
        sex_n.append(1)

        
# removing sex columns 
inputs_n = new_df.drop(['Sex','Survived'],axis='columns')
# adding sex column with bit digit those value add in sex_n array
inputs_n["Sex"] = sex_n
```

```python
inputs_n
```

```python
target = df.Survived
inputs_n['Age'] = Age_n
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_n[['Pclass','Age','Fare','Sex']],target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```

```python
p = model.predict(inputs_n[['Pclass','Age','Fare','Sex']])
```

```python
inputs_n['Survived_n'] = p
```

```python
# Email Spam Exercise
esdf = pd.read_csv('emailSpam.csv')
```

```python
esdf
```

```python
esdf.isnull().sum()
```

```python
Category_n = []
for i in esdf.Category:
    if i == "ham":
        Category_n.append(1)
    elif i == "spam":
        Category_n.append(0)
print(Category_n)
len(Category_n)
```

```python
esdf['Category_n'] = Category_n
esdf
```

```python
X_train, X_test, y_train, y_test = train_test_split(esdf.Message,esdf.Category_n,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

```python
# When we get the data that's certain frequenct then we will use Mutlinomial Naive Bayes
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
email = ['Go until jurong point, crazy.. Available only in bugis n great world la e buffet... Cine there got amore wat...']
email_count = v.transform(email)
model.predict(email_count)
```

```python
X_test_count = v.transform(X_test)
print(f'Training data score {model.score(X_train_count,y_train)} Testing Data Score {model.score(X_test_count,y_test)}')
```

```python
Category_n_predict = v.transform(esdf.Message)
p = model.predict(Category_n_predict)
```

```python
esdf['Category_n_p'] = p
```
