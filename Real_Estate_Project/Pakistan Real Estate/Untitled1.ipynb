```python
df1 = pd.read_csv('Pakistan_estate.csv')
df1.head()
```

```python
df2 = df1.drop(['property_id','location_id','location','purpose','date_added','agency','agent','page_url'],axis='columns')
df2.head()
```

```python
df2['Total_Area'] = df2['Total_Area'].apply(np.int64)
df2
```

```python
Total_Area_n = []
for i in df2.Total_Area:
    value = float(i)
    Total_Area_n.append(value)
df2['Total_Area_n'] = Total_Area_n
df3 = df2.drop(['Total_Area'],axis='columns')
df3
```

```python
df3['price_per_area'] = df3['price']/df3['Total_Area_n']
df3
```

```python
df = df3[(df3.baths < 5) & (df3.property_type == 'House')]
df
```

```python
def plot_scatter_chart(df,pt):
    Flat1 = df[(df.property_type==pt) & (df.bedrooms == 2)]
    Flat2 = df[(df.property_type==pt)& (df.bedrooms == 3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(Flat1.Total_Area_n,Flat1.price_per_area,color='blue',label='2 Bedroom', s=50)
    plt.scatter(Flat2.Total_Area_n,Flat2.price_per_area,marker='+',color='green',label='3 Bedroom', s=50)    
    plt.xlabel('Total Area')
    plt.ylabel('Price Per Area')
    plt.title(pt)
    plt.legend()
plot_scatter_chart(df,'House')
```

```python
property_type_n = []
city_n = []
province_n = []
for pt,c,p in zip(df3.property_type,df3.city,df3.province_name):
    if pt == 'Flat':
        property_type_n.append(1)
    elif pt == 'House':
        property_type_n.append(2)
    elif pt == 'Penthouse':
        property_type_n.append(3)
    elif pt == 'Farm House':
        property_type_n.append(4)
    elif pt == 'Room':
        property_type_n.append(5)
    elif pt == 'Lower Portion':
        property_type_n.append(6)
    elif pt == 'Upper Portion':
        property_type_n.append(7)
    

    if c == 'Islamabad':
        city_n.append(1)
    elif c == 'Karachi':
        city_n.append(2)
    elif c == 'Lahore':
        city_n.append(3)
    elif c == 'Rawalpindi':
        city_n.append(4)
    elif c == 'Faisalabad':
        city_n.append(5)
    
    if p == 'Islamabad Capital':
        province_n.append(1)
    elif p == 'Punjab':
        province_n.append(2)
    elif p == 'Sindh':
        province_n.append(3)
```

```python
df4 = df3.drop(['property_type','city','province_name'],axis='columns')
df4.head()
```

```python
df4['city_n'] = city_n
df4['poperty_type_n'] = property_type_n
df4['province_name'] = province_n
df4.head()
```

```python
X = df4.drop(['price','price_per_area'],axis='columns')
y = df4.price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
lr_clf = LogisticRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```
