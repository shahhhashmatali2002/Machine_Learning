```python
import pandas as pd
df = pd.read_csv('salaries.csv')
df
```

```python
inputs = df.drop('salary_more_then_100k',axis='columns')
target = df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
inputs.head()
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
inputs_n.head()
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n[['company_n','job_n','degree_n']],target)
# model score
model.score(inputs_n[['company_n','job_n','degree_n']],target)
```

```python
model.predict([[2,2,0]])
```

```python
p = model.predict(inputs_n[['company_n','job_n','degree_n']])
```

```python
inputs_n['salary_more_then_100k_n'] = p 
```

```python
inputs_n
```

```python
t_df = pd.read_csv('titanic.csv')
t_df.head()
```

```python
t_df_n = t_df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
t_df_n
```

```python
t_df_n.isnull().sum()
```

```python
import math
median_Age = math.floor(t_df_n.Age.median())
```

```python
median_Age
```

```python
new_t_df = t_df_n.fillna(method="ffill")
new_t_df
```

```python
sex_n = []
for i in new_t_df.Sex:
    if i == 'male':
        sex_n.append(0)
    elif i == 'female':
        sex_n.append(1)

        
# removing sex columns 
inputs_n = new_t_df.drop(['Sex','Survived'],axis='columns')
# adding sex column with bit digit those value add in sex_n array
inputs_n["Sex"] = sex_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
target = t_df_n[['Survived']]
```

```python
model.fit(inputs_n[['Pclass','Age','Fare','Sex']],target)
```

```python
p = model.predict(inputs_n[['Pclass','Age','Fare','Sex']])
```

```python
inputs_n['Survived_n'] = p
inputs_n
```

```python
model.score(inputs_n[['Pclass','Age','Fare','Sex']],target)
```
